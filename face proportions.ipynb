## Converting categorical to numerical

```python
df=pd.get_dummies(df,columns=['Race','Gender'])
```

```python
df
```

## Spliting and training

```python
from sklearn.model_selection import train_test_split

Y = df['Attractive']  
X = df.drop(['Attractive'],axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

## XGBoost Regressor

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_squared_log_error
from math import sqrt
model = XGBRegressor(n_estimators=1000, learning_rate=0.3)
model.fit(X_train, Y_train, 
             early_stopping_rounds=10, 
             eval_set=[(X_test, Y_test)], 
             verbose=False)

pred1 = model.predict(X_test)
```

```python
df1= pd.DataFrame({'Actual': Y_test, 'Predicted': pred1,'Diff':abs(Y_test-pred1)})
df1
```

```python
print("Maximum difference: ",df1['Diff'].max())
print("Minimum difference: ",df1['Diff'].min())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
print("mae: ", mean_absolute_error(Y_test, pred1))
print("rmse: ", sqrt(mean_squared_error(Y_test, pred1)))
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)
pred2 = reg.predict(X_test)
df2 = pd.DataFrame({'Actual': Y_test, 'Predicted': pred2,'Diff':abs(Y_test-pred2)})
df2
```

```python
print("Maximum difference: ",df2['Diff'].max())
print("Minimum difference: ",df2['Diff'].min())
```

```python
print("mae: ", mean_absolute_error(Y_test, pred2))
print("rmse: ", sqrt(mean_squared_error(Y_test, pred2)))
```
